```python
import itertools
import numpy as np
import pandas as pd

import plotly.data as pdata
df = pdata.tips()

def calc_stats(df: pd.DataFrame, calc_colnames: list[str], agg_methods: list) -> pd.DataFrame:
    def calc_stats(group):
        df_ = df.groupby(group)[calc_colnames].agg(agg_methods).reset_index()
        df_.columns = group + stats_result_colnames
        return df_
    return calc_stats

def make_combinabion_list(lst: list) -> list:
    combs = []
    for n in range(1,len(lst)+1):
        for comb in itertools.combinations(lst, n):
            combs.append(list(comb)) #タプルをリスト型に変換
    return combs

grouping_colnames = ['sex', 'day']
calc_stats_colnames = ['total_bill', 'tip']
stats_methods = ['count', 'mean', 'median', 'std']
stats_result_colnames \
    = [calc_col + '_' + method for calc_col, method in itertools.product(calc_stats_colnames, stats_methods)]

# 全体の集計（SeriesだとconcatしにくいのでDummyグループを作成して集計）
df['Dummy'] = 1

groups = make_combinabion_list(['Dummy'] + grouping_colnames)
method = calc_stats(df, calc_stats_colnames, stats_methods)
calc_stats_results = list(map(method, groups))
pd.concat(calc_stats_results, axis=0) \
    .reset_index(drop=True)[grouping_colnames + stats_result_colnames] \
    .replace(np.nan, 'all')
```

```python
import math
import numpy as np
import pandas as pd

import plotly.data as pdata
df = pdata.tips()

df = df.loc[:, ['sex', 'total_bill', 'tip']]

calc_target_col = 'total_bill'
bin_colname = 'bin_' + calc_target_col
min = df[calc_target_col].min()
max = df[calc_target_col].max()
interval = 5
bins = [round(x, 3) for x in np.arange(math.floor(min)-interval, math.ceil(max)+interval*2, interval)]

# ビン分割
df_count_by_bin = pd.cut(df.loc[:, calc_target_col], bins=bins, right=False).sort_index()
df_count_by_bin = df_count_by_bin.rename(bin_colname).reset_index()

# ソート用の順番を示す列を追加（インデックス解除後、unstack時に並びかわってしまうため）
df_cat_total_bill = pd.DataFrame(df_count_by_bin['bin_total_bill'].cat.categories.sort_values(), columns=['bin_total_bill'])
df_cat_total_bill['no_bin_total_bill'] = range(1, len(df_cat_total_bill.index) + 1)
df_count_by_bin = pd.merge(df_count_by_bin, df_cat_total_bill, \
                           left_on='bin_total_bill', right_on='bin_total_bill', how='left')
df_count_by_bin = df_count_by_bin[['bin_total_bill', 'no_bin_total_bill']]

df = pd.concat([df, df_count_by_bin], axis=1)
df[bin_colname] = df[bin_colname].astype(str)

# ------------
calc_target_col = 'tip'
bin_colname = 'bin_' + calc_target_col
min = df[calc_target_col].min()
max = df[calc_target_col].max()
interval = 1
bins = [round(x, 3) for x in np.arange(math.floor(min)-interval, math.ceil(max)+interval*2, interval)]

# ビン分割
df_count_by_bin = pd.cut(df.loc[:, calc_target_col], bins=bins, right=False)
df_count_by_bin = df_count_by_bin.rename(bin_colname).reset_index()

# ソート用の順番を示す列を追加（インデックス解除後、unstack時に並びかわってしまうため）
df_cat_tip = pd.DataFrame(df_count_by_bin['bin_tip'].cat.categories.sort_values(), columns=['bin_tip'])
df_cat_tip['no_bin_tip'] = range(1, len(df_cat_tip.index) + 1)
df_count_by_bin = pd.merge(df_count_by_bin, df_cat_tip, left_on='bin_tip', right_on='bin_tip', how='left')
df_count_by_bin = df_count_by_bin[['bin_tip', 'no_bin_tip']]

df = pd.concat([df, df_count_by_bin], axis=1)
df[bin_colname] = df[bin_colname].astype(str)

grouped = df[['sex', 'no_bin_tip', 'bin_tip','no_bin_total_bill', 'tip']] \
            .groupby(['sex', 'no_bin_tip', 'bin_tip','no_bin_total_bill'])
df_multi = grouped.agg(['count', 'mean', 'median', 'std']).sort_index()

for col in df_multi.columns:
    df_ = df_multi[[col]].unstack().replace(np.nan, '')
    display(df_)

display(df_cat_tip)
display(df_cat_total_bill)
```
