```python
import itertools
import numpy as np
import pandas as pd

import plotly.data as pdata
df = pdata.tips()

def calc_stats(df: pd.DataFrame, calc_colnames: list[str], agg_methods: list) -> pd.DataFrame:
    def calc_stats(group):
        df_ = df.groupby(group)[calc_colnames].agg(agg_methods).reset_index()
        df_.columns = group + stats_result_colnames
        return df_
    return calc_stats

def make_combinabion_list(lst: list) -> list:
    combs = []
    for n in range(1,len(lst)+1):
        for comb in itertools.combinations(lst, n):
            combs.append(list(comb)) #タプルをリスト型に変換
    return combs

grouping_colnames = ['sex', 'day']
# 正しく変換されているか確認すること
df[grouping_colnames] = df[grouping_colnames].replace(np.nan, 'unknown')

calc_stats_colnames = ['total_bill', 'tip']
stats_methods = ['count', 'mean', 'median', 'std']
stats_result_colnames \
    = [calc_col + '_' + method for calc_col, method in itertools.product(calc_stats_colnames, stats_methods)]

# 全体の集計（SeriesだとconcatしにくいのでDummyグループを作成して集計）
df['Dummy'] = 1

groups = make_combinabion_list(['Dummy'] + grouping_colnames)
method = calc_stats(df, calc_stats_colnames, stats_methods)
calc_stats_results = list(map(method, groups))
pd.concat(calc_stats_results, axis=0) \
    .reset_index(drop=True)[grouping_colnames + stats_result_colnames] \
    .replace(np.nan, 'all')
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
sns.set()

import plotly.data as pdata
df = pdata.tips()

y_lim_max = df['tip'].max() + 1
y_lim_min = df['tip'].min() - 1
plt.rcParams["figure.figsize"] = (10, 10)
# 中央値が黒色の実線で、平均値が緑色の破線

# 曜日ごとに分けない
fig, ax = plt.subplots()
plt.ylim(y_lim_min, y_lim_max)
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))

sns.stripplot(data=df, y='tip', alpha=0.5, ax=ax)
sns.boxplot(data=df, y='tip', ax=ax,
  showmeans=True, meanline=True,
  whiskerprops={'visible': False},
  showfliers=True, showbox=False, showcaps=False)
sns.stripplot(data=df, x='day', y='tip', alpha=0.5, ax=ax, color='purple')
sns.boxplot(data=df, x='day', y='tip', ax=ax,
  showmeans=True, meanline=True,
  whiskerprops={'visible': False},
  showfliers=True, showbox=False, showcaps=False)


# 曜日ごとに分ける
for dim in df['day'].unique():
    df_ = df[df['day']==dim]
    fig, ax = plt.subplots()
    
    sns.stripplot(data=df_, y='tip', alpha=0.5, ax=ax)
    sns.boxplot(data=df_, y='tip', ax=ax,
    showmeans=True, meanline=True,
    whiskerprops={'visible': False},
    showfliers=True, showbox=False, showcaps=False)

    sns.stripplot(data=df_, x='sex', y='tip', alpha=0.5, ax=ax, color='purple')
    sns.boxplot(data=df_, x='sex', y='tip', ax=ax,
    showmeans=True, meanline=True,
    whiskerprops={'visible': False},
    showfliers=True, showbox=False, showcaps=False)

    ax = plt.gca()
    ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))

    plt.ylim(y_lim_min, y_lim_max)
    plt.title(dim)
    # plt.savefig(dim + '.png')
    plt.show()

print(y_lim_max)
print(y_lim_min)
```
